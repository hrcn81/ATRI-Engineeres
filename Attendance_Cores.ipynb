```python
import csv
import os
from datetime import date

# ==========================================
# Employee List
# ==========================================

employees = [
    "Pritesh",
    "Kirit",
    "Priyanka",
    "Pratham",
    "Rahul",
    "Khushbu",
    "Raj Gabani",
    "Jaimin",
    "Mihir",
    "Shubham_1",
    "Sushir",
    "Mayank",
    "Akash",
    "Shubham_2",
    "Maulik",
    "Divya",
    "Manisha",
    "Rajendra",
    "Parin",
    "Annie",
    "Riya",
    "Rohan",
    "Vivek",
    "Shubham_3",
    "Suraj",
    "Khushi"
]

FILE_NAME = "attendance_cores.csv"
TODAY = str(date.today())
from google.colab import drive
drive.mount('/content/drive')
FILE_NAME = "/content/drive/MyDrive/attendance_cores.csv"

# ==========================================
# Status Mapping
# ==========================================

status_map = {
    "A": "Absent",
    "L": "Leave",
    "W": "WFH",
    "H": "Half Day",
    "WO": "Week Off",
    "W1": "Half Day + Week Off",
    "W2": "Week Off + Half Day"
}

# ==========================================
# Create CSV
# ==========================================

if not os.path.exists(FILE_NAME):
    with open(FILE_NAME, "w", newline="") as f:
        writer = csv.writer(f)
        writer.writerow(["Date", "Employee", "Status"])

# ==========================================
# Prevent Duplicate Entry
# ==========================================

"""with open(FILE_NAME, "r") as f:
    reader = csv.DictReader(f)

    for row in reader:
        if row["Date"] == TODAY:
            print(f"\nAttendance already marked for {TODAY}")
            exit() """

# ==========================================
# Header
# ==========================================

print("\n" + "=" * 60)
print("📋 DAILY ATTENDANCE MANAGEMENT")
print("=" * 60)

print(f"📅 Date : {TODAY}")
print(f"👥 Employees : {len(employees)}")

print("\nOnly enter employees who are NOT Present.")
print("Press ENTER without typing a name to finish.\n")

print("--------------------------------------------")
print("Status Codes")
print("--------------------------------------------")
print("A   = Absent")
print("L   = Leave")
print("W   = Work From Home")
print("H   = Half Day")
print("WO  = Week Off")
print("W1  = Half Day + Week Off")
print("W2  = Week Off + Half Day")
print("--------------------------------------------")

attendance = {}

# ==========================================
# Attendance Input
# ==========================================

while True:

    name = input("\nEmployee : ").strip()

    if name == "":
        break

    match = None

    for emp in employees:
        if emp.lower() == name.lower():
            match = emp
            break

    if match is None:
        print("❌ Employee not found.")
        continue

    if match in attendance:
        print("⚠ Already entered.")
        continue

    status = input("Status : ").upper().strip()

    if status not in status_map:
        print("❌ Invalid Status.")
        continue

    attendance[match] = status_map[status]

    print(f"✅ {match} → {status_map[status]}")

# ==========================================
# Auto Present
# ==========================================

for emp in employees:
    attendance.setdefault(emp, "Present")

# ==========================================
# Save CSV
# ==========================================

with open(FILE_NAME, "a", newline="") as f:

    writer = csv.writer(f)

    for emp in employees:
        writer.writerow([TODAY, emp, attendance[emp]])

# ==========================================
# Summary Counts
# ==========================================

summary = {
    "Present": 0,
    "Absent": 0,
    "Leave": 0,
    "WFH": 0,
    "Half Day": 0,
    "Week Off": 0,
    "Half Day + Week Off": 0,
    "Week Off + Half Day": 0
}

for status in attendance.values():
    summary[status] += 1

# ==========================================
# WhatsApp Report
# ==========================================

print("\n")
print("=" * 60)
print("📲 WHATSAPP REPORT")
print("=" * 60)

print(f"\n📅 Date : {TODAY}\n")

print("🚫 Non-Present Employees\n")

non_present = False

for emp in employees:

    status = attendance[emp]

    if status != "Present":
        non_present = True
        print(f"• {emp:<15} ➜ {status}")

if not non_present:
    print("Everyone is Present 🎉")

print("\n--------------------------------------------")

print("📊 Summary")
print("--------------------------------------------")

print(f"👥 Total Employees : {len(employees)}")
print(f"✅ Present         : {summary['Present']}")
print(f"❌ Absent          : {summary['Absent']}")
print(f"🌴 Leave           : {summary['Leave']}")
print(f"🏠 WFH             : {summary['WFH']}")
print(f"🕒 Half Day        : {summary['Half Day']}")
print(f"📅 Week Off        : {summary['Week Off']}")
print(f"🌓 W1              : {summary['Half Day + Week Off']}")
print(f"🌗 W2              : {summary['Week Off + Half Day']}")

print("\n✅ Attendance saved successfully.")

# ==========================================
# WhatsApp Copy Format
# ==========================================

print("\n" + "=" * 60)
print("COPY BELOW TO WHATSAPP")
print("=" * 60)

print(f"📅 Attendance - {TODAY}\n")

for emp in employees:

    status = attendance[emp]

    if status != "Present":
        print(f"🔸 {emp} - {status}")

print("\n📊 Summary")
print(f"👥 Total : {len(employees)}")
print(f"✅ Present : {summary['Present']}")
print(f"❌ Absent : {summary['Absent']}")
print(f"🌴 Leave : {summary['Leave']}")
print(f"🏠 WFH : {summary['WFH']}")
print(f"🕒 Half Day : {summary['Half Day']}")
print(f"📅 Week Off : {summary['Week Off']}")
print(f"🌓 W1 : {summary['Half Day + Week Off']}")
print(f"🌗 W2 : {summary['Week Off + Half Day']}")
```
